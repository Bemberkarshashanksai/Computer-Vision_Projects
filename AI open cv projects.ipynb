```python
#CODE
import cv2
import time
import imutils

cam=cv2.VideoCapture(0)
time.sleep(1)

firstFrame=None #we assinged certain flaf to identify which is first frame
area=500
#count=0

while True:
    _,img=cam.read()
    text="Normal"
    img=imutils.resize(img,width=500)
    grayImg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    gaussianImg=cv2.GaussianBlur(grayImg,(21,21),0)
    if firstFrame is None:
        firstFrame=gaussianImg
        continue
    imgDiff=cv2.absdiff(firstFrame,gaussianImg) #image differance between first and next image
    threshImg=cv2.threshold(imgDiff,25,255,cv2.THRESH_BINARY)[1]
    threshImg=cv2.dilate(threshImg,None,iterations=2) #if we have any noise in threshold apply dilation to remove
    cnts=cv2.findContours(threshImg.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE) #to find nearby pixel values
    #we are using only the copy of image,if we use original we cant use it further
    cnts=imutils.grab_contours(cnts) #to grab the contour values
    for c in cnts:
        if cv2.contourArea(c) < area:
            continue
        (x,y,w,h)=cv2.boundingRect(c) #to get values of x,y,w,h
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        text="Moving Object detected"
    print(text)
    #count+=1
    #print(count)
    cv2.putText(img,text,(10,20),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),2)
    cv2.imshow("cameraFeed",img)
    #cv2.imshow("threshimg",threshImg)
    key=cv2.waitKey(1) & 0xFF
    if key==ord("q"):
        break
cam.release()
cv2.destroyAllWindows()      
```

```python
import cv2
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
cam=cv2.VideoCapture(0)
while True:
    _,img=cam.read()
    #text='No person detected'
    grayImg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face=faceCascade.detectMultiScale(grayImg,1.3,4)
    for (x,y,w,h) in face:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        #text='person detected'
    #cv2.putText(img, text, (10,20),cv2.FONT_HERSHEY_SIMPLEX,1,(2,255,1),2)    
    cv2.imshow("person",img)
    key=cv2.waitKey(10) & 0xFF
    if key==ord("a"):
        break
cam.release()
cv2.destroyAllWindows()
    
```

```python

#code
import cv2
import os
datasets = 'dataset'  
sub_data = 'sai'     

path = os.path.join(datasets, sub_data)
if not os.path.isdir(path):
    os.makedirs(path)

(width,heigth)=(130,100)
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
cam=cv2.VideoCapture(0)
count=1
while count<31: #iterations
    #print(count)
    _,img=cam.read()
    grayImg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face=faceCascade.detectMultiScale(grayImg,1.3,4) #to detect only face frame from whole background
    for (x,y,w,h) in face:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        faceOnly=grayImg[y:y+h,x:x+w]
        resizeImg=cv2.resize(faceOnly,(width,heigth))
        cv2.imwrite('%s/%s.png' % (path,count), resizeImg)
    count+=1
    cv2.imshow('face detection',img)
    key=cv2.waitKey(10)
    if key==27:
        break
print("Dataset obtained successfully")

cam.release()
cv2.destroyAllWindows()
    
```

```python
#code
import imutils
import cv2
redLower=(0,193,98) #(H,S,V) lower
redUpper=(179,255,255)

camera=cv2.VideoCapture(0)
while True:
    (grabbed,frame)=camera.read()
    frame=imutils.resize(frame,width=600)
    blurred=cv2.GaussianBlur(frame,(11,11),0) #smoothning
    hsv=cv2.cvtColor(blurred,cv2.COLOR_BGR2HSV)
#applying the colours what we want
    mask=cv2.inRange(hsv,redLower,redUpper) #to mask only for specific color
    mask=cv2.erode(mask,None,iterations=2) #erotion
    mask=cv2.dilate(mask,None,iterations=2) #dilation
    cnts=cv2.findContours(mask.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[-2]
    center=None
    if len(cnts)>0: #to identify the colour....if it identifies any object the cnts increses
        c=max(cnts,key=cv2.contourArea) #max area
        ((x,y),radius) =cv2.minEnclosingCircle(c) #drawing min enclosing circle
        M=cv2.moments(c) #with the help of moments we can able to find the center which helps to drawn min enc circle
        center=(int(M["m10"] / M["m00"]), int(M["m01"] / M["m00"])) #finds the centef of area
        if radius >10:
            cv2.circle(frame,(int(x),int(y)),int(radius),(0,255,255),2) #it changes acc to object,its dynamic
            cv2.circle(frame,center,5,(0,0,255),-1) #to draw a dot for center
            print(center,radius)
            if radius >250:
                print("stop")
            else:
                if(center[0]<150): #we checking the x cordinate of center
                    print("Left")
                elif(center[0]>450):
                    print("Right")
                elif(radius<250):
                    print("Front")
                else:
                    print("Stop")
    
    cv2.imshow("Farme",frame)
    key=cv2.waitKey(1) &0xFF
    if key==ord("q"):
           break
    
camera.release()
cv2.destroyAllWindows()
#this is the tecnique used by nasa in 2020 to cetect craters on the mars
```
